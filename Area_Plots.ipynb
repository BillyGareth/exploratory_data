```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.rand(10,4),columns=list('abcd'))
```

```python
df.plot.area()
# stacked by default and missing values filled by 0
```

```python
df.plot.area(stacked=False)
```
