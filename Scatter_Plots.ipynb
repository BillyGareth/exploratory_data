```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.rand(50,4),columns=list('abcd'))
```

```python
df.plot.scatter(x='a',y='b')
```

```python
ax=df.plot.scatter(x='a',y='b',color='DarkBlue',label='Grroup1')
df.plot.scatter(x='c',y='d',color='DarkGreen',label='Group2',ax=ax)
```
