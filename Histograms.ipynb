```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range('2020-11-01',periods=1000))
ts=ts.cumsum()
```

```python
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list('ABCD'))
df=df.cumsum()
```
