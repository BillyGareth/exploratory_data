```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range('2020-11-20',periods=1000))
```

```python
ts=ts.cumsum()
```

```python
ts.plot()
```

```python
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list('ABCD'))
```

```python
df=df.cumsum()
df.plot()
```

```python
df3=pd.DataFrame(np.random.randn(1000,2),columns=['B','C']).cumsum()
```

```python
df3['A']=pd.Series(list(range(len(df))))
```

```python
df3.plot(x='A',y='B')
```
